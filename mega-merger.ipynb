```python
from pymote.algorithms.mega_merger import MegaMerger
```

```python
%matplotlib inline

from pymote import NetworkGenerator
from pymote import write_pickle, read_pickle
from pymote.environment import Environment2D

net_gen = NetworkGenerator(10, comm_range=500)
net = net_gen.generate_random_network()
net.environment = Environment2D(shape=(800,800))
net.show()

net.algorithms = (MegaMerger, )

write_pickle(net, "MegaMerger.npc.gz")
```

```python
from pymote import Simulation

sim = Simulation(net)

sim.reset()
sim.run()
```

```python
for node in net.nodes():
    print(node.id,
          node.memory["level"],
          node.memory["downtown_id"],
          node.memory["city_name"],
          node.memory["neighbours_in_city"]
     )
```
